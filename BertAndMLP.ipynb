```python
def getencode(texts, debug=False):
    tokens =[ tokenizer.tokenize(text) for text in texts]
    if debug:
        print("[tokens]",tokens)
    indices, segments = [],[]
    for text in texts:
        idx,seg = tokenizer.encode(first=text, max_len=512)
        if debug:
            print("idx,",idx[:len(text)+2])
            print("seg,",seg[:len(text)+2])
        indices.append(idx)
        segments.append(seg)
    print(np.array(indices).shape, np.array(segments).shape)
    predicts = model.predict([np.array(indices), np.array(segments)])
    return predicts
```
